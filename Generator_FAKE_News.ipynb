## Загрузка модели


```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
DEVICE = torch.device("cuda:0")

# Для большей наглядности будем работать с русскоязычной версией GPT от Сбера.
# Ниже находятся команды для загрузки и инициализации модели и токенизатора.
model_name = "sberbank-ai/rugpt3medium_based_on_gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name).to(DEVICE)
```

```python
from transformers import TextDataset, DataCollatorForLanguageModeling

# Путь к .txt файлу
train_path = "/content/text.txt"

# Создание датасета
train_dataset = TextDataset(tokenizer=tokenizer, file_path=train_path, block_size=64)

# Создание даталодера (нарезает текст на оптимальные по длине куски)
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
```

## Обучение

Для файнтюнинга нам понадобится объект класса `Trainer`, который сделает всю грязную работу за нас. Необходимо всего лишь запустить метод `trainer.train()`.

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="./finetuned",       # директория с выходными данными
    overwrite_output_dir=True,      # перезаписываем содержимое выходной директории при каждом запуске
    num_train_epochs=10,           # число эпох обучения
    per_device_train_batch_size=32, # batch size для обучения
    per_device_eval_batch_size=32,  # batch size для выполнения
    warmup_steps=10,                # количество шагов для "прогрева" (управление скоростью обучения)
    gradient_accumulation_steps=16, # накопление градиента (16 шагов накапливаем градиенты для batch_size, эмуляция вычисления на пакете 16 * 32 для слабых GPU)
    report_to="none"
    )

# Инициализируем класс обучения
trainer = Trainer(
    model=model,                 # модель
    args=training_args,          # параметры обучения
    data_collator=data_collator, # загрузчик данных
    train_dataset=train_dataset, # датасет для обучения
    optimizers = (torch.optim.AdamW(model.parameters(), lr=1e-5), None) # оптимизатор
)
```

Запускаем обучение:

```python
trainer.train()
```

Сэмплирование с Ограничением Маловероятных Токенов

```python
# Функция для генерации фейковых новостей
def generate_fake_news(model, tokenizer, prompt, max_length=200, min_length=50):
    # Кодируем исходную подсказку в тензор для модели
    input_ids = tokenizer.encode(prompt, return_tensors="pt").to("cuda:0")

    # Генерация текста с контролем повторений и завершения
    out = model.generate(
        input_ids,
        do_sample=True,
        num_beams=6,                  # Используем beam search для улучшения качества генерации
        temperature=0.2,               # Пониженная температура для меньшей случайности
        top_k=50,                     # Увеличиваем top_k, чтобы модель могла выбирать из более широкого спектра слов
        top_p=0.9,                    # Повышаем top_p для лучшего разнообразия
        max_length=max_length,       # Ограничиваем длину текста
        min_length=min_length,       # Минимальная длина текста
        eos_token_id=tokenizer.eos_token_id,  # Используем правильный токен для завершения
        early_stopping=True,          # Останавливаем генерацию на первом достижении окончания
        repetition_penalty=2.0,       # Штрафуем за повторение фраз
    )

    # Декодируем результат в текст
    generated_text = tokenizer.decode(out[0], skip_special_tokens=True)


    # Проверяем, что текст не слился с другим. Можно добавить дополнительные проверки на "переход" текста.
    # Например, если длина текста превышает 1 новость, можно обрезать до первой точки или восклицательного знака.
    if '.' or '!' in generated_text:
        generated_text = generated_text.split('.')[0] + '.'

    # Возвращаем сгенерированный текст
    return generated_text
```

```python
# Пример генерации фейковых новостей
prompts = [
    "Эксперты утверждают, что",
    "В Москве произошло необычное событие: ",
    "Сенсационное открытие в области науки: ",
    "Новости с фронта: ",
    "Новости из Санкт-Петербурга: ",
    "Последние новости из Беларусии: ",
    "Последние слухи из мира политики: ",
    "Cобытия в спортивной сфере: ",
    "Тайные расследования: ",
    "Важные новости из мира экономики: "
]
```

```python
# Генерируем несколько фейковых новостей
generated_news = []
for prompt in prompts:
    fake_news = generate_fake_news(model, tokenizer, prompt)
    generated_news.append(fake_news)
```
